# Introduction

How does microbial ecosystem structure and functioning respond to warming and disturbance (e.g. chemical pollution such as pesticides or antibiotics)? 

We develop a mathematical model to investigate the response of ecosystems/communities of interacting bacterial populations to address this. 

Here we derive and describe the model, and analyze it with the objective of determining how systems of multiple interacting bacteria with different functional traits (e.g. size, type of response to warming, or the thermal performance curves (TPC) resource preference) are self-organized or "selected" by species sorting under different environment conditions. We are particularly interested in which bacterial traits/strategies maximize ecosystem functioning (e.g., carbon flux, biomass production, or total respiration), and which consistently emerge through interaction-driven assembly and turnover dynamics. 

# The Model

We build on / modify the mathematical model proposed by [Marsland et al. (2019)](https://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1006793), which is itself based on the seminal Rosenzweig-MacArthur consumer-resource equations. 

The model consists of coupled ODEs representing one or more consumers (bacteria) feeding on one or more carbon resources. The resources are substrates in the from of particulate or dissolved organic carbon substrates (POC and DOC, respectively). Each bacterial population can feed on a specific set of resources, and a portion of any given resource consumed by any given bacterial population is necessarily converted into one or more other resources during the process of sequestration and / or allocation to maintenance and growth by the population. 

We begin by deriving the multi-species community/ecosystem model from fundamental processes and assumptions. 

## A Basic Population Growth Model

We start with the fact that the carbon biomass growth rate of the $i^{th}$ bacterial population biomass ($C_i$) must be given by the gain of carbon ($U$) minus loss ($D$) rates (both in units of (carbon) mass $\times$ time$^{-1}$):

\begin{equation}\label{eq:pop1}
    \frac{1}{C_i}\frac{dC_i}{dt} = U_i - D_i
\end{equation}

Note that $U$ as such is not a scalar, but a function that depends on the availability of growth-limiting nutrients such as N and P, which we assume henceforth to be abundantly available in the growth medium. For example, $U$ may be of the Michaelis-Menten form $U = \frac{\mu_{max}}{K+N}$, where $\mu_{max}$ is nutrient-unlimited, maximal uptake and growth rate, $K$ is the half-saturation constant, and $N$ is the limiting nutrient. Then, $\lim_{N\to\infty} U \to \mu_{max}$, i.e., effectively, $U = \mu_{max}$.

We can separate the loss term $D$ into carbon lost due to respiration $R$ (i.e. as $\text{CO}_2$), and that lost/released/leaked back into the environment/medium as organic substrate ($L$) as well as cell mortality ($z$):

\begin{equation*}
    \frac{1}{C_i}\frac{dC_i}{dt} = U_i - R_i - L_i - z_i
\end{equation*}

We can further partition the respiratory loss $R$ into the cost of maintenance ($R_m$) and that which occurs as a result of growth $R_g$ (Manzoni & Porporato ([2009](https://www.sciencedirect.com/science/article/pii/S0038071709000765))), i.e., 

\begin{equation} \label{eq:R}
    R = R_g + R_m
\end{equation}

Growth respiration here represents the loss of carbon due to the thermodynamic limits (inefficiency) to processing and assimilating carbon sources ([Gommers et al. 1988](https://onlinelibrary.wiley.com/doi/abs/10.1002/bit.260320112)). Maintenance respiration represents the costs of processes such as protein and nucleic acid turnover, osmo-regulation and cell movement ([Chapman & Grey 1986](https://www.sciencedirect.com/science/article/pii/0038071786900957)). In theory, separating these two types of respiration is desirable, for two reasons. First, in cells that are not in any sort of stationary phase, $R_m$ is arguably independent of carbon uptake such that carbon loss from maintenance is proportional to the amount of metabolically active biomass, because maintenance activities continue irrespective of whether carbon uptake is occurring. Second, as growth respiration rate is associated with assimilating/allocating carbon into biomass, it should be related to the magnitude of uptake (e.g., $R_g = f(U) = \varepsilon U$, where $\varepsilon$ is a proportion representing growth or carbon allocation efficiency). However, in practice, differentiating between these two types of respiratory carbon losses is impossible, both, because these processes effectively occur simultaneously in the ODE system, and are indistinguishable using standard empirical/laboratory experiments.    

In reality, what is quantifiable is Carbon Use Efficiency, which we will denote henceforth as $p_g$. This is typically estimated empirically from growth rate measures coupled with respirometry as:   

\begin{equation*}
    p_g = \frac{\mu_{max}}{\mu_{max} + R},
\end{equation*}

where $R$ is as defined in eqn \ref{eq:R}, and we use $U$ instead of $\mu_{max}$ because uptake and growth under nutrient-unlimited carbon are essentially indistinguishable empirically.  Including these assumptions into the population growth model (eqn \ref{eq:pop1}) gives:

\begin{equation}
    \frac{1}{C_i} \frac{dC_i}{dt} = U_i - p_{g,i} U_i - L_i - z_i
\end{equation}

Similarly we can link the non-respiratory carbon loss $L_i$ to the processes occurring within the cell. As this loss term is mostly due to the inefficiency of metabolizing carbon uptake (with the contribution of cell mortality being negligible), we can also link it to carbon uptake, assuming a fixed proportion is lost via this route $L = p_L U$:

\begin{array}
\frac{1}{C_i}\frac{dC_i}{dt} &= U_i - p_{g,i} U_i - p_L U_i - z_i\\
                                 &= U_i (1 - p_{g,i} - p_L) - z_i\\
\end{array}


## Multiple carbon sources

The above equation represents the growth of a population with respect to some arbitrary rate of carbon uptake $U$. Real populations can be resource-limited (leading to a dependence of uptake on resource availability) and generally feed on multiple carbon sources. We can represent the uptake of carbon by the $i^{th}$ species feeding on $M$ carbon sources ($S$) as:

\begin{equation*}
U_i = \sum_{j=1}^M u_{ij} S_j
\end{equation*}

Here $u_{ij}$ (units of $(\text{mass} \times \text{time})^{-1}$) is the per-unit biomass uptake rate of the $j^{th}$ resource. Note that this rate is described by a scalar here, but can be a function of resource availability (such as Type II "functional response").

As the growth respiration $R_g$ rates and the loss of carbon to the environment $L$ are dependent on uptake, we can assume that their values will vary across the $M$ carbon substrates. We can indicate this by rewriting the population growth equation as:

\begin{equation*}
    \frac{1}{C_i}\frac{dC_i}{dt} = \left(\sum_{j=1}^M u_{ij} S_j  (1 - p_{g,j} - p_{L,j})\right) - z_i
\end{equation*}

Furthermore, loss of carbon to the environment should actually occur via the release of carbon in the form of multiple substrates due to the breakdown of more complex substrates. This can be expressed as sum of losses across $M$ carbon substrates:

$$p_{L,j} = \sum_{k=1}^M l_{jk},$$ 

where entries of the matrix $l_{jk}$ represent the proportion of uptake of resource $j$ that is released as resource $k$. These "leakages" allow the commonly-seen cross-feeding microbial interactions to be represented in the dynamics. Thus, the final growth model for the $i^{th}$ population becomes:

\begin{equation}
    \frac{1}{C_i}\frac{dC_i}{dt} = \left(\sum_{j=1}^M u_{ij} S_j  (1 - p_{g,j} - \sum_{k=1}^{M} l_{jk})\right) - z_i
\end{equation}

## Resource dynamics

Now that we have derived an equation describing the growth of a bacterial species, we need to write the set of equations describing resource dynamics. These equations follow fairly naturally from the population dynamics by enforcing mass conservation. As with bacterial carbon biomass, the amount of the $j^{th}$ resource  in the environment must be determined by the rate at which it enters ($I$) minus the rate at which it leaves ($D$):

\begin{equation}
   \frac{dS_j}{dt} = I_j - D_j
\end{equation}

Variation in resource input $I$ can be attributed either to abiotic processes (which we represent here through the constant term $\rho_j$) or through the activity of the bacterial community which takes up the $j^{th}$ resource at the rate $U_j$ and releases it (via the inefficiency of uptake) through the $L_j$ term:

\begin{equation}
    \frac{dS_j}{dt} = \rho_j - U_j + L_j
\end{equation}

We can get the rates of uptake and release by the community by summing across all $N$ species in the community and across all $M$ resources which they release as a result of the $L_j$ term:

\begin{equation}
    \frac{dS_j}{dt} = \rho_j - \sum_{i = 1}^{N} \left(u_{ij} S_j C_i   - \sum_{k = 1}^{N} u_{ik} l_{kj} S_k C_i \right)
\end{equation}

So overall the change in environmental carbon in the form of resource $j$ is equal to the intrinsic rate of change ($\rho_j$) minus the net effect of each bacterial population. This net effect is made up of the uptake (the first term) and the release via the inefficient uptake of other carbon sources (the last).

# The final, general model

Thus, the full model is:

\begin{equation}
    \frac{dC_i}{dt} = C_i\left(\sum_{j = 1}^M u_{ij} S_j\Bigg(1 - p_{g,j} - \sum\limits_{\substack{k = 1 \\ j \neq k}}^M l_{jk}\Bigg) - z_i \right)
\end{equation}

\begin{equation}
    \frac{dS_j}{dt} = \rho_j - \sum_{i = 1}^{N} \left(u_{ij} S_j C_i  - \sum_{j \neq k}^M u_{ij} l_{kj} S_j C_i \right) 
\end{equation}

Here $C_i$ is the biomass of consumer $i$, $u_{ij}$ is the preference of consumer $i$ to uptake resource $j$, $S_j$ is the mass or concentration of resources $j$, $p_{g,j}$ is the loss of consuming resource $j$ to respiration, $l_{jk}$ is the proportion of $j$ taken up by the bacteria that is converted to and released as resource $k$, $R_{m,i}$ is the maintenance cost loss term, $\rho$ is the external input of resource $j$ and $l_{kj}$ is the gain and conversion of resource $j$ from $k$.

## Comparison to the Marsland model

The  Marsland et al (2019) model is:

\begin{equation}
\frac{dC_i}{dt} = g_i C_i\left(\sum_{j=1}^{M} w_j \sigma u_{ij} S_j\Bigg(1 -\sum_{k=1}^{M}l_{jk}\Bigg) - R_{m,i}\right)
\end{equation}

\begin{equation}
\frac{dS_j}{dt} = \rho_j + \sum_{i=1}^{N}C_i\left(\sum_{k=1}^{M}\sigma u_{ij}S_j l_{kj} - \sigma u_{ij}S_j\right) \tag{4}
\end{equation}

Here, $g_i$ is the proportional constant between per capita growth rate and energy flux, $w_j$ is the energy density of resource $j$ (units are energy/mass) and $\sigma$ is the functional response (units mass/time).

The differences between the current and Marsland's model are:

1. We do not consider energy density of resource ($w_j$) nor the constant between growth and energy flux ($g_i$) 
* We assume a Type I functional response ($\sigma = 1$)
* For Marsland $\rho_j$ = 0 except for $\rho_0$, so only the first resource has an external input which then diversifies. In our model all resources have an input.
* We include the addition of respiration loss associated with growth ($R_g$) as a loss term in the Bacterial mass (dC/dt) equation.

## Parameterizations and Constraints

So there are 4 main parameters that we need for this model:
1. Uptake matrix
2. Maintenance Respiration
3. Growth Respiration
4. Carbon transition matrix

We now consider the issue of obtaining values or estimates for each of these parameters.

## Uptake

So we cannot really obtain estimates of uptake directly but we can may be able to use Tom Smith's data to get estimates of the magnitude relative to total respiration?

In terms of structure of the matrix we can vary the nestedness of the interactions (changing the strength of interaction in the community).

Temperature should directly affect uptake as it is related to the activity of enzymes and the transport of substrates across the cell wall, at a minimum the rate of passive diffusion should increase. It's hard to know how to parameterise this individually though... There is some evidence that microbes switch metabolic pathways with temperature, this can be represented though the alteration of the relative uptake rates. 

## Maintainence Respiration

There is a fairly large body of work considering the importance of microbial maintenence respiration with various models and experiments aiming to understand its dynamics.

We may be able to estimate the amount of respiration that originates from this process in relation to the total respiration of a given population by using these papers. For example [Price & Sowers (2004)](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC384798/) estimate the ratio of metabolic rates for bacteria in growth and maintence phases to be $10^3 : 1$. We need to look at more papers for this as this ratio seems very high.

The same paper also looks at the temperature dependence of metabolism across these two phases, finding no difference. This could be very helpful as it indicates that the total respiratory temperature dependence should be equal to the temperature dependence of each part individually...

## Growth respiration

As it stands in the model, growth respiration is the fixed proportion of each carbon source that is lost as $CO_2$ when it is metabolised. Thus, it takes the form of a vector with length $M$. In theory we can set a limit to the values in this vector based on the fact that the metabolism of many carbon sources will not allow total conversion of carbon mass. This should hold true for more complex carbon substrates though the proportion respired should increase as the molecules get more simple (i.e. i imagine the consumption of something simple like glucose should be relatively straightforward and efficient). 

This point is backed up by the work of [Gommers et al. (1988)](https://onlinelibrary.wiley.com/doi/abs/10.1002/bit.260320112) who estimate that around $12\%-29\%$ of substrate is released as $CO_2$ during metabolism (A nice range to work with?).

In terms of temperature dependence the proportion of carbon respired per unit input should not change with temperature (i.e. $p_g$ above) though the supply of material to metabolise (via the uptake term) will. 

## $l$ - the substrate transition matrix

Though this matrix is impossible to directly parameterize, we can make some assumptions about its structure.

Firstly we can assume that its entries are independent of temperature. As this parameter represents the conversion of the uptake of one form of carbon into others it will be constrained by the biochemical pathways that are used in the metabolism of the carbon source. Thus, unless the metabolic pathways change themselves (which there actually is some evidence for ): ) the entries of $l$ will remain the same.

We can also make some assumptions about the structure of $l$. As the transitions between carbon sources represent biochemical reactions we can assume some kind of directionality in the reactions as complex substrates are broken down into more simple ones. This can be interpreted in terms of the entropy production in these reactions.

The following table lists the constraints on the parameters and variables (nutrients and bacteria). 


| Parameter/Variable |Symbol| Matrix/Vector dimension | Units | Constraint | Temperature/Size-dependence | Initial value |
|---|---|---|---|---|---|---|
|Bacterial (Consumer) abundance                   |$C_i$      |   vector $(i,)$ | mass         | >0  | NA | 0.1 |
|Resource abundance                               |$R_j$      | vector $(j,)$ | mass         | >0  | NA | 1  |
|External supply of $j^{th}$ resource             |$\rho_j$ | vector $(,j)$ | mass $\times$ time$^{-1}$     | >0  | no | 1 |
|Uptake of $j^{th}$ Resource by $i^{th}$ consumer |$u_{ij}$ | matrix $(i,j)$ | (mass $\times$ time)$^{-1}$ |>0 | yes | 1 |
|Growth respiration of $j^{th}$ consumer          |$p_{g,j}$   | vector $(,j)$ | 1          | <1  | no | 0  |
|Maintenance respiration of $i^{th}$ consumer     |$R_{m,i}$   | vector $(i,)$ | time$^{-1}$        | >0, <1| yes | 0.5 |
|Excretion of $j^{th}$ to $k^{th}$ resource       | $l_{jk}$ | matrix $(j,k)$ | 1  | <1  | no | 0.4 |

There is an additional condition: that growth respiration and excretion must not sum to 1 or more ($p_{g,j} + l_{jk} < 1$), because otherwise no carbon is gained by the bacteria.


# Analysis

To gain some mechanistic/analytical insights, we begin the analysis with simple 2 bacterial population systems. 


## 2-species 2-resources 

The system is:

\begin{equation}
    \frac{dC_1}{dt} = C_1 \left((u_{11} S_1 (1 - Rg_1 - (l_{12} + l_{21})) + u_{12} S_2 (1 - Rg_2 - (l_{21} + l_{12}))) - R_{m,1} \right)
\end{equation}

\begin{equation}
   \frac{dC_2}{dt} = C_2 \left((u_{21} S_1 (1 - Rg_1 - (l_{12} + l_{21})) + u_{22} S_2(1 - Rg_2 - (l_{21} + l_{12}))) - R_{m,2} \right) 
\end{equation}

\begin{equation}
    \frac{dS_1}{dt} = \rho_1 - \left((C_1 u_{11} S_1 - (C_1 u_{11} S_1 l_{21} + C_1 u_{11} S_1 l_{12})) + (C_2 u_{21}S_1-   (C_2 u_{21} S_1 l_{21} + C_2 u_{21} S_1 l_{12}))\right)
\end{equation}

\begin{equation}
    \frac{dS_2}{dt} = \rho_2 - \left((C_1 u_{12} S_2 - (C_1 u_{12} S_2 l_{12} + C_1 u_{12} S_2 l_{21})) + (C_2 u_{22} S_2 - (C_2 u_{22} S_2 l_{12} + C_2 u_{22} S_2 l_{21}))\right)
\end{equation}

## Modelling functional traits

### Temperature and size dependance

In this model only $u_{ij}$ and $R_{m,i}$ can be temperature- and size-dependant as they are the only terms that explicitly depend on an individual consumer $i$. Temperature- and size-dependancy are formulated following the Metabolic Theory of Ecology and the Sharpe-Schoolfield equation. Such that:


\begin{equation}
u_{ij} = B_{u}M^{-0.25} * \frac{e^{\frac{-Ea}{k}\cdot\left(\frac{1}{T} - \frac{1}{T_{ref}}\right)}}{1 + \frac{Ea}{E_D - Ea}\cdot e^{\frac{-E_D}{k}\cdot\left(\frac{1}{T_{pk}} - \frac{1}{T}\right)}} \tag{11}
\end{equation}

\begin{equation}
R_{m,i} = B_{rm}M^{-0.25} * \frac{e^{\frac{-Ea}{k}\cdot\left(\frac{1}{T} - \frac{1}{T_{ref}}\right)}}{1 + \frac{Ea}{E_D - Ea}\cdot e^{\frac{-E_D}{k}\cdot\left(\frac{1}{T_{pk}} - \frac{1}{T}\right)}} \tag{12}
\end{equation}

where $BM^{-0.25}$ is the mass ($M$) scaled to $3/4$ / M = $M^{-0.25}$ multiplied by a constant that is higher for uptake ($u_{ij}$) than maintenance respiration ($R_{m,i}$). The second part of the equation is the Sharpe-Schoofield equation where $Ea$ is the activation energy in eV, $E_D$ is the deactivation energy after the temperature maximum metabolism is reached ($T_{pk}$), $T$ is the temperature and $T_{ref}$ is the reference temperature. All temperatures are expressed in Kelvin (K) and $k$ is the Boltzmann constant of $8.62 x 10^{-5} eV K^{-1}$.

```python
import Vector_Testing_model as vtm
import matplotlib.pylab as plt
from matplotlib.lines import Line2D
from scipy.integrate import odeint
import numpy as np
import scipy as sc

# Set up parameters
N = 10
M = 10
t_fin = 100
k = 0.0000862
Tref = 273.15 # 0 degrees C
pk = 20 # 
T_pk = Tref + pk
B_g = 0.5 # B0 for growth (uptake)
B_rm = 0.1 # B0 for mainternace respirations
Ma = 1 # Mass
T = 273.15 + 20
Ea = np.concatenate([np.repeat(0.6,N/2), np.repeat(1.0,N/2)])
Ea_D = np.repeat(3.5,N) # Deactivation energy
t_fin = 100
t = sc.linspace(0,t_fin-1,t_fin)
U = vtm.params(N, M, T, k, Tref, T_pk, B_g, B_rm,Ma, Ea, Ea_D)[0]
Rm = vtm.params(N, M, T, k, Tref, T_pk, B_g, B_rm,Ma, Ea, Ea_D)[1] 
Rg = vtm.params(N, M, T, k, Tref, T_pk, B_g, B_rm,Ma, Ea, Ea_D)[2] # l + Rg must be less than 1
l = vtm.params(N, M, T, k, Tref, T_pk, B_g, B_rm,Ma, Ea, Ea_D)[3]
p = vtm.params(N, M, T, k, Tref, T_pk, B_g, B_rm,Ma, Ea, Ea_D)[4]
l_sum = np.sum(l, axis=1)
x0 = np.concatenate((sc.full([N], (0.1)),sc.full([M], (1.0))))
pars = (U, Rm, Rg, l, p, l_sum, Ea, Ea_D, N, M, T, Tref, B_rm, B_g, Ma)

# Run model
pops = odeint(vtm.metabolic_model, y0=x0, t=t, args = pars)
print(pops[t_fin-1,:])

#### Plot output ####
t_plot = sc.linspace(0,100,100)
plt.plot(t, pops[:,0:N], 'g-', label = 'Consumers', linewidth=0.7)
plt.plot(t, pops[:,N:N+M], 'b-', label = 'Resources', linewidth=0.7)
plt.grid
plt.ylabel('Population density')
plt.xlabel('Time')
plt.title('Bacteria-Nutrients population dynamics')
plt.legend([Line2D([0], [0], color='green', lw=2), Line2D([0], [0], color='blue', lw=2)], ['Bacteria', 'Nutrients'])
plt.show()
```

#### Incorporating temperature

The solutions for bacteria at equilibrium where the Arrhenius equation is used (i.e. no Tpk):

\begin{equation}
C = \frac{\rho(1 - Rg - l)}{B_{Rm}M^{-0.25}e^{(-Ea/k)(\frac{1}{T} - \frac{1}{T_{ref}})}
(1-l)} \tag{13} 
\end{equation}

and for resources are:

<span class="mark">\begin{equation}
S = \frac{B_{rm}M^{-0.25}e^{(-Ea/k)(\frac{1}{T} - \frac{1}{T_{ref}})}}{B_{u}M^{-0.25}e^{(-Ea/k)(\frac{1}{T} - \frac{1}{T_{ref}})}(1-Rg-l)} =  \frac{B_{rm}}{B_{u}(1-Rg-l)}\tag{14}
\end{equation}

Resource ($S$) is independant of $Ea$ and individual functional traits aside from $B0$</span>

#### Find all possible equilibria Bacteria masses at different Ea values

Rearranging Equation 11 yields:

\begin{equation}
Ea = \frac{kln\left(\frac{\rho(1-Rg-l)}{CB_{Rm}M^{-0.25}(1-l)}\right)}{\frac{1}{T} - \frac{1}{T_{ref}}} \tag{15}
\end{equation}

```python
k = 0.0000862; Rg = 0; l = 0.4; T = 293.15; Tref = 273.15; Ma = 1; B_rm = 0.1; p=1

C = sc.linspace(0.001,10,100) # Bacteria
Ea = -(k * np.log((p * (1-Rg-l))/ (C * B_rm * Ma**-0.25 * (1-l))))/((1/T)-(1/Tref)) # Ea

from sklearn import linear_model
lm = linear_model.LinearRegression()
model = lm.fit(Ea.reshape(-1,1),np.log(C)) # run linear model

Ea1 = sc.linspace(0,4,100) # predict for plotting, Ea (x values)
C1 = np.exp(lm.intercept_ + (Ea1 * lm.coef_)) # predict for plotting, C (y values)

# Plot
plt.plot(Ea1, C1, 'g-', label = 'Consumers', linewidth=1.7)
plt.grid
plt.ylabel('Bacteria Biomass')
plt.xlabel('Activation Energy')
plt.show()
```

Bacteria with higher Eas have lower mass, unless competing...see below.

#### Anayltical by sympy

```python
from sympy import *
import scipy as sc
u1, u2, S, Rg, Rm1, Rm2, C1, C2,rho, t = var("u1, u2, S, Rg, Rm1, Rm2, C1, C2, rho,t",real = True)

dC1_dt = C1 * ((u1 * S * (1-Rg)) - Rm1)
dC2_dt = C2 * ((u2 * S * (1-Rg)) - Rm2)
dS_dt = rho - ((u1 * S * C1) + (u2 * S * C2))
dC1_dt, dC2_dt, dS_dt
C1_eqlb = Eq(dC1_dt, 0)
C2_eqlb = Eq(dC2_dt, 0)
S_eqlb = Eq(dS_dt, 0)
C1_eqlb,C2_eqlb, S_eqlb
C1_eqlb_sol = solve(C1_eqlb, S)
C2_eqlb_sol = solve(C2_eqlb, S)
S_eqlb_sol = solve(S_eqlb, C1)
print(C1_eqlb_sol);print(C2_eqlb_sol); print(S_eqlb_sol); print(solve((C1_eqlb,C2_eqlb, S_eqlb), C1,C2,S))
```

## R*

To determine which species will win we can calculate the $R*$, the amount of resource needed for a species to remain at equilibrium. To do this we insert the equilbrium equation for bacteria ($C$, Eq 9) into the equation for bacteria ($C$, Eq 1) and solve for S:

\begin{equation}
0 = \frac{uS\rho(1 - Rg - l)(1-Rg-l)}{Rm(1-l)} - Rm 
\end{equation}

\begin{equation}
uS\rho(1 - Rg - l)^2 = Rm^2(1-l) 
\end{equation}

\begin{equation}
R*  = \frac{Rm^2(1-l)}{u\rho(1 - Rg - l)^2} \tag{18}
\end{equation}


Both nutrient uptake ($u$) and maintenance respiration ($Rm$) can be used to define functional groups based on their temperature dependancies.


- When T = T_pk the difference in R* between different individuals is largest.
- Groups with lower $Ea$ have competitive advantage
- The assembly model (below) selects for species with low $Ea$

### Expressing size dependancy

It is uncertain whether size would increase fitness/population size in bacteria. This is because a larger size could increase genome increasing growth rate, but could also increase maintenance respiration ($R_{m,i}$) which would negatively impact on growth rates and population size. Instead of forcing size to have a positive or negative impact on population size both these relationships can be incorporated into the model. As body size increases, $u_{ij}$ will rise, increasing uptake and have a positive impact on population growth. $R_{m,i}$ will also increase with body size, but this is a loss term from the bacteria consumer so decreases overall population growth. $B$ for $u_{ij}$ (growth, $Bg$) will be higher than for $R_{m,i}$ (respiration, $Brm$).
